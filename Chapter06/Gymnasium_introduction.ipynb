```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML


fig, ax = plt.subplots()
ax.axis('off')  # Turn off axes

patch = ax.imshow(frames[0], animated=True)

def animate(i):
    patch.set_data(frames[i])

anim = animation.FuncAnimation(fig, animate, frames=len(frames), interval=100)

HTML(anim.to_jshtml())
```

```python
import gymnasium as gym
from gymnasium.utils.save_video import save_video
env = gym.make("Pong-v4", render_mode="rgb_array_list")
_ = env.reset()
step_starting_index = 0
episode_index = 0
for step_index in range(5000):
   action = env.action_space.sample()
   _, _, terminated, truncated, _ = env.step(action)
   #print(step_index, terminated, truncated)
   if terminated or truncated:
      save_video(
         env.render(),
         "videos",
         fps=env.metadata.get("render_fps", 30),
         step_starting_index=step_starting_index,
         episode_index=episode_index
      )
      step_starting_index = step_index + 1
      episode_index += 1
      env.reset()
env.close()
```
