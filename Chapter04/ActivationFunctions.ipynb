```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of the sigmoid function
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Generate a range of values
x = np.linspace(-10, 10, 100)

# Plotting the sigmoid function and its derivative
plt.figure(figsize=(8, 4))
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, sigmoid_derivative(x), label="Derivative", color='orange')
plt.title("Sigmoid Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Tanh function
def tanh(x):
    return np.tanh(x)

# Derivative of the tanh function
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

# Generate a range of values
x = np.linspace(-10, 10, 100)

# Plotting the tanh function and its derivative
plt.figure(figsize=(8, 4))
plt.plot(x, tanh(x), label="tanh")
plt.plot(x, tanh_derivative(x), label="Derivative", color='orange')
plt.title("Tanh Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Derivative of the ReLU function
def relu_derivative(x):
    return np.where(x > 0, 1, 0)

# Generate a range of values
x = np.linspace(-10, 10, 200)

# Plotting the ReLU function and its derivative
plt.figure(figsize=(8, 4))
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, relu_derivative(x), label="Derivative", color='orange')
plt.title("ReLU Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Softmax function
def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum()

# Simplified derivative of the softmax function
def softmax_derivative(softmax_values):
    return softmax_values * (1 - softmax_values)  # Simplified for one variable

# Generate a range of values and apply softmax
x = np.linspace(-10, 10, 100)
softmax_values = softmax(x)

# Calculate the simplified derivative
derivative_values = softmax_derivative(softmax_values)

# Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, softmax_values, label="Softmax")
plt.plot(x, derivative_values, label="Simplified Derivative", color='orange')
plt.title("Softmax Function and Simplified Derivative")
plt.xlabel("Input value (x)")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky ReLU function
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

# Derivative of the Leaky ReLU function
def leaky_relu_derivative(x, alpha=0.01):
    return np.where(x > 0, 1, alpha)

# Generate a range of values
x = np.linspace(-10, 10, 200)

# Plotting the Leaky ReLU function and its derivative
plt.figure(figsize=(8, 4))
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.plot(x, leaky_relu_derivative(x), label="Derivative", color='orange')
plt.title("Leaky ReLU Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ELU function
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

# Derivative of the ELU function
def elu_derivative(x, alpha=1.0):
    return np.where(x > 0, 1, alpha * np.exp(x))

# Generate a range of values
x = np.linspace(-10, 10, 200)

# Plotting the ELU function and its derivative
plt.figure(figsize=(8, 4))
plt.plot(x, elu(x), label="ELU")
plt.plot(x, elu_derivative(x), label="Derivative", color='orange')
plt.title("ELU Function and its Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.grid()
plt.legend()

plt.show()
```
